```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Loading the data 
df = pd.read_csv('all_data.csv')
print(df.head())
df
```

```python
## Here I will see What is the average life expectancy in these nations?
plt.figure(figsize=(10, 6)) 
sns.barplot(x='Country', y='Life expectancy at birth (years)', data=df)

# Add labels and a title
plt.xlabel('Country')
plt.ylabel('Average Life Expectancy at Birth (years)')
plt.title('Average Life Expectancy by Country')

# Show the plot
plt.show()
plt.clf
```

```python
## The distribution of the life expectancy above
sns.histplot(data=df, x='Life expectancy at birth (years)', bins=15, kde=True, color='blue')

# Add labels and a title
plt.xlabel('Life Expectancy at Birth (years)')
plt.ylabel('Frequency')
plt.title('Distribution of Life Expectancy')

# Show the plot
plt.show()
```

```python
## Here i will see how life expectancy increased over time in the six nations?

#plt.figure(figsize=(10, 6))  
sns.lineplot(x='Year', y='Life expectancy at birth (years)', hue='Country', data=df)

# Add labels and a title
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (years)')
plt.title('Changes in Life Expectancy Over Time')

# Show the plot
plt.show()
```

```python
## Here i will see how GPD has icreased over time in the six nations
sns.lineplot(x='Year', y='GDP', hue='Country', data=df)

# Add labels and a title
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('Changes in GDP Over Time')

# Show the plot
plt.show()
```

```python
## Here i will see if there is a coorelation between the GDP and life expectancy of a country 

correlation_coefficient = df['GDP'].corr(df['Life expectancy at birth (years)'])
print(f'Correlation Coefficient: {correlation_coefficient}')

plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Life expectancy at birth (years)'], alpha=0.7)
plt.xlabel('GDP')
plt.ylabel('Life Expectancy at Birth (years)')
plt.title('Scatter Plot of GDP vs. Life Expectancy')
plt.grid(True)
plt.show()
```
